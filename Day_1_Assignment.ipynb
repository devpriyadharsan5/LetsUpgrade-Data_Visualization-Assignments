days = [1,2,3,4,5,6,7] #days of d week

**Question 3**

 Create a 3 by 3 subplots:

multiple plots

x = [1,2,3,4]

y1 = [4,3,2,1]

y2 = [10,20,30,40]

y3 = [40,30,20,10]

y4 = [1,2,1,2]

y5 = [40,70,90,70]


```python

x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplots(figsize=(15,15))


plt.subplot(3,3,1)
plt.plot(x,y1,'r--')

plt.subplot(3,3,2)
plt.plot(x,y2,'g--')

plt.subplot(3,3,3)
plt.plot(x,y3,'r--')

plt.subplot(3,3,4)
plt.plot(x,y4,'p--')

plt.subplot(3,3,5)
plt.plot(x,y5,'b--')
```
